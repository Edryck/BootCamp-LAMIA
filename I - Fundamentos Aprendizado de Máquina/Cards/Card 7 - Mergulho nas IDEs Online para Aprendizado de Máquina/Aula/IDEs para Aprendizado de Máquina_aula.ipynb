# Mergulho nas IDEs Online para Aprendizado de Máquina

Um breve resumo sobre algumas IDEs para Aprendizado de Máquina, vamos falar sobre o Python Shell, IPython, Notebooks de Jupyter (Jupyter Notebook, Jupyter Lab), Google Colab e os Notebooks do Kaggle.

## Python Shell
Se já está programando em Python, deve saber que é possível digitar `python` em qualquer terminal do computador e será levado ao Python Shell, nele você estará podendo executando um linha de código linha por linha, logo, não é muito recomendável fazer isso.

## IPython
É nele que os notebooks do Jupyter são construidos, o IPython é semelhante ao Python, mas tem mais funcionalidades. ELe foi desenvolvido como uma extensão do Python para tornar o Python mais interativo e os notebooks de Jupyter foram construídos com base nele.

## Jupyter
É necesssário sua instalação, caso contrário não será possível utilizar ele. Nessa seção estarei falando tanto do Jupyter Notebook quanto do Jupyter Lab, para sua instalação abrindo o terminal digite:

### Jupyter Notebook  
É um ambiente onde pode salvar, anotar coisas (acredito que seja óbvio por ser chamado "Notebook"), carregar, executar arquivos Python ou no próprio Jupyter (`.ipynb`), ele é baseado em cells (células) tendo dois tipos principais:  
- **Markdown:** Basicamente é para a linguagem de marcação leve para formatar textos simples, muito usado para criar documentos técnicos, por ser baseado em HTML é possível fazer formatações usando HTML nele, por exemplo, as observações que eu deixo pequena ao decorrer desse documento.
- **Code:** São cells (células) de código, onde podemos escrever Python, onde o código estará sendo executando em um ambiente Python kernel em segundo plano. Podemos realizar os *imports* e também podemos realizar importações de imagens ou gráficos e ele será exibido e renderizado.  
A junção dessas duas funcionalidades das células se torna possível escrever textos juntos dos códigos assim fazendo uma documentação ou explicação sobre cada trecho do código.

**Exemplos:**

```python
print('Hello World!')
```

```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib
```

```python
# Dados para plotagem
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Tempo (s)', ylabel='Voltagem (mV)',
       title='O mais simples possível')
ax.grid()
```

### JupyterLab  
É um ambiente um pouco diferente do Jupyter Notebook. Ele inclui a possibilidade de adicionar temas e extensões, ele conta com uma guia na lateral com: 
- ***File Browser:*** Possibilita navegar pelos arquivos que temos. Assim tornando mais eficiente e fácil trabalhar com vários arquivos.
- ***Running Terminals and Kernels:*** Ele permite visualizar todos os notebooks, abas que estão abertas e kernels ou instâncias Python que estão em execução. Se clicar em `Shut Down All`, ele vai parar os kernels em segundo plano, se executar alguma célula em seguida, ele vai iniciar outro kernel.  
- ***Table of Contents:*** É uma aba que mostra o seu índice, isso ajuda a procurar e pular de um índice de forma mais rápida.
- ***Extensions:*** Essa permite instalar diversas extensões.

No JupyterLab também é possível exportar o notebook em diferentes tipos de formatos, como: HMTL, LaTeX, Markdown, PDF, Webpdf, entre outros. Exportando em HTML e PDF irá permitir que uma pessoa que não tenha o Jupyter instalado, possa visualizar o notebook.  

<small>**OBS.:** Uma coisa interessante dos notebooks é que não é necessário realizar a execução das células em ordem, porém, isso seria uma má prática, principalmente no caso em que você for reiniciar o kernel e executar todas as células novamente, se um trecho precisar de um *import* e ele estiver depois desse trecho, é certeza que vai dar erro.</small>

## Google Colab
O JupyterLab e Notebook se tornou tão popular que agora é possível executar em instâncias de nuvem também. Uma das maneira de executar o Jupyter na nuvem é utilizar o Google Colab e o Kaggle.  
O interressante de executar na nuvem é que não é necessário a instalação do Python na máquina local nem os pacotes que iria precisar, o ambiente normalmente já é iniciado com diversos pacotes de Ciência de Dados.  
**Diferenças:**  
Por mais que seja parecido, há algumas diferenças, como:  
O Markdown é chamado de Text.  
Ele mostra informações como quanto da instância você está utilizando.  
Ele é salvo direto no Drive e também é possivel copiar para o GitHub.  
Com certeza é possível baixar o arquivo `.ipynb` ou `.py`.  


## Kaggle  
O Kaggle é um site gratuito que usa muitas das mesmas funcionalidades do Jupyter, claro que para isso é necessário criar um conta nele.  
Uma das coisas legais de trabalhar no ambiente Kaggle é que já te permite utilizar todos o *datasets* públicos que eles fornecem, no quais é possível vincular nos Notebooks e carregar, para isso é só ir em `Add data`, nele vai mostrar todos os *datasets* públicos disponíveis, também é possível adicionar alguns filtros. 
Para usar o `read_csv` do Pandas, por exemplo, para passar o caminho do arquivo, normalmente ele é salvo no diretório de entrada(*input*), ficaria assim o caminho:  
`../input/nome-do-dataset/nome-do-arquivo.csv`  
Assim você consegue iniciar sua análise de dados.  
Diferente do Jupyter, o Kaggle te força a executar o código do inicio ao fim para salvar.  
Outra coisa interessante é que você pode compartilhar o código do notebook publicamente, assim qualquer pessoa com o link poderá visualizá-lo.
