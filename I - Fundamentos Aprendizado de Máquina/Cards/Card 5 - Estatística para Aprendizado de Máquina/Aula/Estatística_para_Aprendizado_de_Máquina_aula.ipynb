## Tipos de Dados  
Os principais tipos de dados são:  
- **Numéricos**.
- **Categóricos**.
- **Ordinal**.

### Numéricos  
Basicamente os dados do tipo numérico são usados para representar algo quantitativo, ou seja, que você pode medir.  
Exemplo: Preços de algo, tempo, etc.  
Dentro dos dados do tipo numérico existem dois tipos:  
- **Dados Discretos**: São baseados em números inteiros, geralmente são contagens de algum evento.
  Exemplos:  
  Quantas compras um cliente fez em um ano?  
  Quantas vezes eu virei a cara?
- **Dados Contínuos**: Ele tem um número infinito de valores possíveis
  Exemplos:
  Quanto tempo levou para um usuário finalizar a compra?
  Quanto choveu ontem?

### Categóricos  
Esses dados não tem um significado matemático característico, eles não é um tipo de dado que dá para comparar com outro de forma direta, o que normalmente fazemos é atribuir algum valor numérico para representar ele, mas esse valor não tem um significado inerente.  
Exemplos: Dados binários, Gênero, Raça, etc.  

### Ordinal  
Uma mistura de dados numéricos e dados categóricos, onde os dados categóricos têm algum significado matemático  
Exemplo: Avaliações de filmes em uma escala de 1 a 5.
As avaliações devem ser 1 a 5 e esses valores tem significado matemático, nesse caso, 1 significa que é um filme péssimo e 2 significa que o filme é melhor que 1, mas pior que 3, etc.  

## Média, Mediana e Moda  
Esse tópico é mais para lembrar sobre o que são esses três, mas também sobre como usar em Python.  

- **Média**: É a simplificação de Média Aritméticas, que nada mais é que a soma de todos os valores dividido pela quantidade de valores.  
  Fórmula: $M = S / Q$;
  Onde M é a média, S é a soma de todos os termos e Q é a quantidade de termos.  
  Exemplo: Média dos números 1, 2, 3, 2, 6, 4, 5. M = (1 + 2 + 3 + 2 + 6 + 4 + 5) / 7 = 23 / 7 = aproximadamente 3,28.
- **Mediana**: Organizando os valores em ordem crescente, a mediana vai ser o valor localizado no meio. No caso da quantidade de valores for par, tem que calcular a média dos dois números que estão no meio.
  Exemplo: Usando os mesmos números da média, (1, 2, 3, 2, 6, 4, 5), se organizar em ordem crescente fica assim: (1, 2, 2, 3, 4, 5, 6) e a mediana vai ser o valor no meio, logo, 3 é a mediana.
- **Moda**: Moda é o valor que mais aparece, usando os mesmos valores anteriormente, o valor que mais aparace é o 2 (aparece duas vezes), logo, a moda é 2.

## Variação e Desvio Padrão  
Variação e Desvio Padrão são medidas de disperção de uma distribuição de dados.  
<small>**OBS.:** Os dados utilizados para os exemplo estão nos slides da aula, mas para simplificar aqui, basicamente é um histograma representando as chegadas de pessoas por minuto.</small>

### Variância  
Basicamente a variância (σ²) é a média dos quadrados das diferenças em relação à média.  
Exemplo: Qual é a variância do conjunto de dados (1, 4, 5, 4, 8)?
<ol>
    <li>
        Calculamos a média: (1 + 4 + 5 + 4 + 8) / 5 = 4,4.
    </li>
    <li>
        Agora pegamos a diferença em relação à média: (-3,4, -0,4, 0,6, -0,4, 3,6).
    </li>
    <li>
        Agora essa diferença elevamos ao quadrado: (11,56, 0,16, 0,36, 0,16, 12,96).
    </li>
    <li>
        Por último, calculamos a média das diferenças ao quadrado: σ² = (11,56, 0,16, 0,36, 0,16, 12,96) / 5 = 5,04.
    </li>
</ol>

### Desvio Padrão  
Basicamente o Desvio Padrão é a raiz quadrada da variância.  
σ² = 5,04  
σ = √5,04  
σ = 2,24  
O desvio padrão de (1, 4, 5, 4, 8) é 2,24.  
Geralmente é usado como uma forma de identificar outliers, ou seja, pontos de dados que estão a mais de um desvio padrão da média podem ser considerados incomuns.
Você pode falar sobre o quão extremo um ponto de dados é falando sobre “quantos sigmas” ele está de distância da média.

### Variância da População vs. Variância da Amostra  
Se você estiver trabalhando com uma amostra de dados em vez de um conjunto de dados inteiro (a população inteira), então deve-se usar a "variância da amostra" em vez da "variância da população".  
Para N amostras, basta dividir as variâncias ao quadrado por N-1 em vez de N.  
No nosso exemplo, calculamos a variância da população assim:  
σ² = (11,56 + 0,16 + 0,36 + 0,16 + 12,96) / 5 = 5,04  
Mas a variância da amostra seria:  
S² = (11,56 + 0,16 + 0,36 + 0,16 + 12,96) / 4 = 6,3

### Fórmula da Variância  
- Variância da População:  
  $σ² = \frac{Σ(X - µ)^2}{N}$

- Variância da Amostra:  
  $S² = \frac{Σ(X - M)^2}{N - 1}$  
  
Onde:  
M e µ é a média.  
X é o conjuntos de valores.  
N é a quantidade de valores.  
S é para a variância da amostra.  
σ é para a variância da população.  

<small>**OBS.:** Para o desvio padrão, a fórmula seria essa: $σ = √\frac{Σ(X - µ)^2}{N}$</small>

## Função de Densidade de Probabilidade e de Massa de Probabilidade  
Já vimos e usamos anteriormente uma função de distribuição normal em alguns exemplos, esses exemplos são exemplos de funções de densidade de probabilidade. Existem outros tipos de funçõesde densidade de probabilidade, na qual vai ser mostrado no curso.  

### Função de Distribuição Normal e Função de Massa de Distribuição   
Já vimos alguns exemplos anteriormente e ela é um exemplo de função de densidade de probabilidade.

````python
# Eu estou criando um array com 10000 número gerados aleatoriamente
# seguindo uma distribuição normal com média 0 e desvio padrão 1
x = np.random.normal(0, 1, 10000) 
# O matplotlib vai pegar esse 10000 valores, verifica qual é o maior e menor
# valor e divide esse intervalo em 100 caixas (o arg bin) iguais
plt.hist(x, 100)
# Eu peguei um valor bem alto (10000 é bastante... eu acho) para mostrar o desenho
# da "curva de sino" de forma mais nítida
plt.show()
````

É mais fácil pensar como a probabilidade de um determinado valor ocorrer, porém é meio enganoso quando se trata de dados contínuos, porque tem  um número infinito de pontos de dados possíveis em uma distribuição contínua (poderia ter 0, 0,001, 0,00001, etc.), portanto a probabilidade de um valor muito específico ocorrer é muito  pequena.  
A função de densidade de probabilidade se refere à probabilidade de uma determinada faixa de valores ocorrerem, por exemplo, em uma distribuição normal, entre a média e um desvio padrão da média, tem 34,1% estar dentro da faixa (-1, 0) e (0, 1), seria melhor explicado com a imagem do vídeo do curso. Da para ampliar ou refinar essa faixa para descobrir os valores reais.  
Basicamente uma função de distribuição de probabilidade vai mostrar a probabilidade de um ponto de dados estar dentro de um determinado intervalo de um determinado valor.  
<small>**OBS.:** A função de dstribuição normal é um exemplo de uma função de distribução de probabilidade</small>  

Quando está  lidando com dados discretos, a nuance de sobre ter um número infinito de valores possíveis desaparece, isso é chamado de forma diferente, que é a função de massa de distribuição, por exemplo, podemos plotar uma função de densidade de probabilidade normal de dados contínuos, seria representado por uma linha em forma de sino no exemplo acima, mas se quantizar isso em um conjunto de dados discretos, como foi feito acima usando um histograma, poderia dizer que o número 0 ocorre pouco mais de 30% de chance. Portanto, a função de massa de probabilidade é a maneira como visualizamos a probabilidade de ocorrência de dados discretos.  

**Diferença de terminologia:**  
- **Função de Densidade de Probabilidade**: Uma curva sólida que descreve a probabilidade de uma faixa de valores ocorrer com dados contínuos.  
- **Função de Massa de Probabilidade**: Probabilidade de ocorrência de determinados valores em um conjunto de dados.

### Exemplos de Distribuições de Dados  
Vou explicar de forma breve os exemplos de distribuições de dados e usando Python.

#### Distribuição Uniforme  
Uma distribuição uniforme significa que tem uma probabilidade constante de um valor ocorrer dentro de um determinado intervalo.

````python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
# Utilizar o uniform para pedir um conjunto de valores aleatórios uniformemente distribuidos
# que variam entre -10 e +10, 100000 desses valores
values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
````

#### Distribuição Normal/Gaussiana  

````python
from scipy.stats import norm
import matplotlib.pyplot as plt
# Cria uma lista de valores para plotar o intervalo de -3 e +3 com um incremento de 0,001 entre eles
x = np.arange(-3, 3, 0.001)
# No eixo Y é uma função de distribuição normal
plt.plot(x, norm.pdf(x))
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
# Gera números aleatórios com uma distribuição normal, como são elementos aleatório então não vair ter a
# curva perfeita como no de cima
mu = 5.0 # Média en torno da qual queremos centralizar os dados
sigma = 2.0 # Desvio padrão desses dados, dispersõa deles
values = np.random.normal(mu, sigma, 10000) # 10000 é a quantidades de pontos de dados
plt.hist(values, 50) # Histograma divido em 50 intervalos
plt.show()
````

#### Exponencial PDF  
O Função de Distribuição de Probabilidade Exponencial é outra função vista com frequência onde os valores decaem exponencialmente.

````python
from scipy.stats import expon
import matplotlib.pyplot as plt
# Cria valores entre 0 e 10 com um incremento de 0,001
x = np.arange(0, 10, 0.001)
# Plota os valores de x no eixo y que é definido pela função de densidade de probabilidade exponencial 
plt.plot(x, expon.pdf(x)) # Função do Scipy.stats para a exponencial
plt.show()
````

#### Função de Massa de Probabilidade Binomial  
Lida com dados discretos, no caso do exemplo está lidando com dados discretos.

````python
from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5
# Cria valores discretos de x de 0 a 10 com um espaçamento de 0,001
x = np.arange(0, 10, 0.001)
# PLota uma função de probabilidade de massa binomial usando os valores de x
plt.plot(x, binom.pmf(x, n, p))
plt.show()
````

#### Função de Massa de Probabilidade de Poisson  
Tem uma aplicação mais específica, exemplo: Temos informações sobre o número médio de coisas que acontecem em um determinado período de tempo, isso pode dar uma maneira de prever as chances de obter um valor diferente em um determinado dia futuro.

````python
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 500 # Média de 500
# Define valores de x para observar entre 400 e 600, com um espaçamento de 0,5
x = np.arange(400, 600, 0.5)
# Plota usando uma função de probabilidade de Poisson
plt.plot(x, poisson.pmf(x, mu))
plt.show()
````

Momentos é um termo matemático para maneiras de medir o formato de uma distribuição de dados, de uma função de densidade de probabilidade ou de qualquer coisa.  
Matematicamente temos algumas notações para defini-los, e você (eu) sabe como eles são definidos se tiver conhecimento em cálculo (acho que tenho, fiz a matéria de cálculo já).  
Notação:  
$µ_n = \int_{-\infty}^{+\infty} (x - c)^n f(x)dx$  
Tradução:  
Tá calculando a diferença de cada valor e o resultado é elevado a n, onde n é o número do momento, e integrando ao longo de toda função, de menos infinito a infinito.  

O primeiro momento é a média do dados que estão sendo analisados.  
O segundo momento é a variância.  
O terceiro e quarto momento, bom... Aqui complica.  
O terceiro momento é chamado de assimetria, uma medida de quão assimétrica é uma distribuição.  
O quarto momento é chamado de curtose, é a espessura da cauda e a nitidez do pico, a medida da forma da distribuição dos dados.  
Os quatro momentos:  
1° Média - a média  
2° Variância - a variância  
3° Assimetria - quão assimetrica é a distribuição  
4° Curtose - a medida da forma dos dados  
Bora pro código agora.

````python
%matplotlib inline
# Criando um conjunto de dados aleatórios com a distribuição normal
import numpy as np # Como eu tô pegando o código do curso vou manter os imports, mas eles não são necessários a essa altura
import matplotlib.pyplot as plt # Eu poderia só colocar lá no topo do arquivo os imports, mas vou deixar assim para não me perder
# ... Novamente, centraliza em zero, média é 0 e desvio-padrão é 0,5 e 10000 pontos de dados
vals = np.random.normal(0, 0.5, 10000)
plt.hist(vals, 50)
plt.show()
````

````python
# Já fizemos isso, o numpy já tem as funções prontas, mas vamos lá
# O primeiro momento - média
np.mean(vals)
````

````python
# Lembrando, o desvio padrão é a raiz da variância, lembra que colocamos o desvio padrão em 0,5?
# Logo a variância tem que ser 0,25
# Segundo momento - variância
np.var(vals)
````

````python
# Para calcular a assimetria vamos precisar do scipy, especificamente o pacote stats
import scipy.stats as sp
# Como os dados estão centrados em 0, a assimetria deve ser quase 0
# Terceiro momento
sp.skew(vals)
````

````python
# Para calcular a curtose também usamos o scipy
# Ela vai descrever o formato da cauda, para uma distribuição normal, ele deve ser próxima de 0
sp.kurtosis(vals)
````

#### Diagrama de Caixa e Bigode  
Visto frequentemente em corretoras, quem já mexeu como bolsa de valores está bem familiarizado.

A linha vermelha representa a mediana dos dados e a caixa representa os limites do primeiro e terceiro quartis.

## Covariância e Correlação  
São maneira de medir se dois atributos diferentes estão relacionados em um conjunto de dados.

#### Covariância  
Mede como duas variáveis variam em conjunto em relação as suas médias.  
Pense nos conjuntos de dados para duas variáveis como vetores de alta dimensão, fazemos para cada atributo, em cada ponto de dados, converta eles em vetores de variâncias em relação à média, depois calcule o produto escalar (cosseno do ângulo entre eles) dos dois vetores, por último divida pelo tamanho da amostra.  
Porém interpretar a covariância é difícil, uma covariância pequena, próxima de 0, significa que não tem muita correlação entre as duas variáveis, já covariâncias grandes, longe de 0 (pode ser negativa para relações inversas), significa que tem uma correlação.  
Mas quão grandes? Aí entre o problema da correlação, ela normaliza tudo pelo desvio padrão de cada atributo. Assim, dá pra afirmar de forma clara que uma correlação negativade um significa uma correlação inversa perfeita, ou seja, à medida que um valor aumenta, o outro diminui e vice-versa. Uma correlação de zero significa que não
há correlação entre os dois conjuntos de atributos, e uma correlação de um implica uma correlação perfeita, onde dois atributos se movem exatamente da mesma forma, independentemente do ponto de dados analisado.  
Portanto, correlação  não implica causalidade. O fato de encontrar um valor de correlação muito alto não significa que um desses atributos causa o outro, mas significa apenas que existe uma relação entre os dois, e essa relação pode ser causada por algo completamente diferente. A única maneira de realmente deterinar uma causalidade é por meio de um experimento controlado.  
### Calculando correlação e causalidade em Python  

````python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from pylab import *

def de_mean(x):
    """
    Função de desvio padrão
    Recebe um conjunto de dados (uma lista) e calcula média desse conjunto da dados
    A sintaxe do return significa que vai criar uma lista, percorrer cada elemento x, chamando
    de xi e então retornar a diferença entre xi e a média xmean
    Essa função retorna uma nova lista de dados que representa os desvios padrão da média para cada ponto de dados
    """
    xmean = mean(x)
    return [xi - xmean for xi in x]


def covariance(x, y):
    """
    Como a covariância é definida
    Ela é definida como um produto escalar (dot()), que é a medida do ângulo entre dois vetores (de_mean())
    de_mean() - Um vetor dos desvios em relação à média para um determinado conjunto de dados
    Depois dividimos por n - 1, porque esta sendo lidado com uma amostra
    """
    n = len(x) 
    return dot(de_mean(x), de_mean(y)) / (n-1)

"""
Para esse exemplo, vai ser criado  alguns dados que tentarão encontrar um relação entre
a velocidade de carregamento da página de um site e quanto as pessoas gastam
"""
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)

plt.show()
````

````python
covariance (pageSpeeds, purchaseAmount) 
# O valor da covariância está próximo de zero, logo não tem relação entre as duas coisas
````

Agora vamos fazer com que o valor da compra seja uma função real da velocidade da página.  
Vai manter as coisas um pouco aleatórias aqui, mas estamos criando um relação real entre esses dois conjuntos de valores. Assim, para um uma determinada tendência, para um determinado usuário, exixte uma relação real entre as velocidades de página que ele encontra e o valor que ele gasta.

````python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)

plt.show()
````

````python
covariance (pageSpeeds, purchaseAmount)
````

Agora dá para ver que tem uma pequena curva, onde as coisas tendem a estar bem alinhadas.  
As coisas ficam um pouco irregulares na parte de baixo, perto da parte inferior, devido à aleatoridade, mas calculando a coveriância, chegamos a um valor muito maior, isso implica uma correlação positivo ou negativo (nesse caso negativa). Ainda é difícil de interpretar o que o 9 realmente significa, é ai que entra a correlação, onde vamos normalizar tudo pelos desvios padrão.

````python
def correlation(x, y):
    # Calculamos a correlação de entre dois conjuntos de atributos (x e y).
    stddevx = x.std() # Calculamos o desvio padrão de cada um
    stddevy = y.std()
    # Calculamos a covariância de cada um (covariance()) e divide pelo desvio padrão de cada um
    return covariance(x,y) / stddevx / stddevy  # Na vida real tem que checar se não vai dividir por 0

# Isso resulta no valor da covariância, que é normalizado para 1 ou -1
correlation(pageSpeeds, purchaseAmount)
````

O valor da correlação entre esses dois elementos resultou em -0,3, isso indica que há alguma correlação entre esses dois elemetos na direção negativa. Não é perfeita, não é uma linha reta perfeita, mas tem algo interessante acontecendo ali.

O NumPy pode calcular a correlação usando a função `corrcoef`, isso para se quiser fazer da maneira mais fácil, o resultado é um array que fornece a correlação entre todas as combinações possíveis dos conjuntos de dados eu vou passar. A maneira de interpretar isso é:  
1 implica que há uma correlação perfeita entre comparar a velocidade da página consigo mesma e o valor da compra consigo mesma, mas quando compara a velocidade com o valor de compra ou vice-versa, chegamos a um valor -0,310, que é aproximadamente o valor que conseguimos quando fizemos o cálculo manualmente.

````python
np.corrcoef(pageSpeeds, purchaseAmount)
````

````python
correlation (pageSpeeds, purchaseAmount)
````

## Probabilidade Condicional  
O conceito é simples, basicamente a probabilidade condicional é que, se eu tenho dois eventos que dependem um do outro, posso fazer uma afirmação sobre probabilidade condicional. A  probabilidade de o segundo evento ocorrer dado que o primeiro evento ocorreu.

#### Notação  
$P(B|A) = \frac{P(A, B)}{P(A)}$  
P(A, B) é a probabilidade de ambos A e B ocorrem.  
P(B|A) é a probabilidade de B dado que A ocorreu.  

Isso implica alguma dependência entra b e a.

#### Exemplo  
Aplicando duas provas aos meus alunos e, no geral, 60% deles tenham passado em ambas. Se chamarmos as provas de A e B, P(A, B) seria 60%. No entanto, a primeira prova foi fácil: 80% dos alunos passaram nela. Então, se B for a segunda prova e A for a primeira, P(A) seria igual a 80%. Aplicando na fórmula, fica assim:  
$P(B|A) = \frac{P(A,B)}{P(A)} = \frac{0,6}{0,8} = 0,75$  
Ou seja, 75% dos estudantes que passaram na primeira, passou na segunda.  
Isso é a probabilidade condicional aplicada.

## Teorema de Bayes  
O teorema de Bayes é a probabilidade de A dado B é igual à probabilidade de A vezes a probabilidade B dado A, dividida pela probabilidade de B.  
- **Notação:**  
$P(A|B) = \frac{P(A)P(B|A)}{P(B)}$  
  
- **Exemplo:** Um teste de drogas, podemos perguntar qual é probabilidade de ser um usuário de uma droga, dado que o teste deu positivo.  
  
O teorema de Baye é importante porque ele destaca que isso depende muito tanto da probabilidade de A quanto da probabilidade de B. Portanto, a probabilidade de ser usuário de uma droga, dado que o teste deu positivo, pode ser muito diferente da probabilidade de testar positivo, dado que você era usuário de drogas. Como você pode ver, existe um problema real: teste diagnósticos em medicina ou teste de drogas produzem muitos falsos positivos. Ainda assim, podemos afirmar que a probabilidade de um teste detectar um usuário pode ser muito alta, mas isso não significa necessariamente que a probabilidade de ser usuário, dado que o teste deu positivo, seja alta.  
São duas coisas diferentes, e o teorema de Bayes permite quantificar essa diferença.  

#### Exemplo  
Criamos um teste de drogas que possa identificar usuários com precisão. O teste de drogas acerta 99% das vezes e dá um resultado negativo para 99% dos não usuários, mas apenas 0,3% da população total uusa a droga questão. Portanto, temos um probabilidade muito pequena de realmente ser um usuário de uma droga.  

Podemos fazer os cálculos. Vamos considerar que o evento A significa que você é usuário de alguma droga e o evento B significa testou positivo para a droga usando este teste. Precisamos calcular a probabilidade geral de um teste positivo, e podemos fazer isso comparando a probabilidade de um teste positivo se você for usuário com a probabilidade de um teste positivo se você não for usuário. Isso resulta em 22,8%, o que equivale a 1,3% neste exemplo.  
Então, temos a probabilidade de B, a probabilidade geral de um teste positivo para a droga sem saber mais nada sobre você. Agora, se fizermos o cálculo, a probabilidade de ser usuário da droga, dado que você testou positivo, é a probabilidade de um resultado positivo no teste, o que significa que você realmente é usuário de drogas. A probabilidade geral de ser um usuário de droga é de 3%. Sabemos que 3% da população usa drogas, multiplicada pela probabilidade de um teste positivo dado que você é um usuário. E, novamente, este teste tem o que parece ser uma altíssima precisão de 99%. Então, temos 0,3% da população usa drogas, multiplicando pela 99% de precisão, dividido pela probabilidade geral de um teste positivo, o que resulta em 1,3%. Portanto, a probabilidade de ser um usário real de droga, dado teste deu positivo, é de apenas 22,8%.  

$P(A|B) = \frac{P(A)P(B|A)}{P(B)} = \frac{0,003 * 0,99}{0,013} = 22,8\% $  

Ou seja, mesmo que este teste de drogas seja preciso me 99% dos casos, ele ainda fornece um resultado falso na maioria das vezes em que o teste dá positivo.
