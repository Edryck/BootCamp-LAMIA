def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n-1)

### Probabilidade Condicional

Abaixo, segue um código para criar dados fictícios sobre a quantidade de coisas que as pessoas compram, considerando sua faixa etária.  
Ele gera 100.000 "pessoas" aleatórias e as classifica aleatoriamente em faixas etárias de 20, 30, 40, 50, 60 ou 70 anos.  
Em seguida, atribui uma probabilidade menor de compra para pessoas mais jovens.   
Ao final, temos dois dicionários em Python:  
`totals` contém o número total de pessoas em cada faixa etária.  
`purchases` contém o número total de itens comprados por pessoas em cada faixa etária.  
O total geral de compras está em totalPurchases, e sabemos que o número total de pessoas é 100.000.  
Vamos executar o código e dar uma olhada:  

```python
random.seed(0)

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchases = 0
for _ in range(100000):
    ageDecade = random.choice([20, 30, 40, 50, 60, 70])
    purchaseProbability = float(ageDecade) / 100.0
    totals[ageDecade] += 1
    if (random.random() < purchaseProbability):
        totalPurchases += 1
        purchases[ageDecade] += 1
```

Vamos brincar com probabilidade condicional.

Primeiro, vamos calcular P(E|F), onde E significa "compra" e F significa "você tem 30 e poucos anos". A probabilidade de alguém com 30 e poucos anos comprar algo é simplesmente a porcentagem de pessoas com 30 e poucos anos que compraram algo:

```python
PEF = float(purchases[30]) / float(totals[30])
print('P(purchase | 30s): ' + str(PEF))
```

P(F) é simplesmente a probabilidade de ser 30 neste conjunto de dados:

```python
PF = float(totals[30]) / 100000.0
print("P(30's): " +  str(PF))
```

E P(E) é a probabilidade geral de comprar algo, independentemente da sua idade:

```python
PE = float(totalPurchases) / 100000.0
print("P(Purchase):" + str(PE))
```

Se E e F fossem independentes, esperaríamos que P(E | F) fosse aproximadamente igual a P(E). Mas não são; P(E) é 0,45 e P(E|F) é 0,3. Portanto, isso nos diz que E e F são dependentes (o que sabemos que são neste exemplo).

P(E,F) é diferente de P(E|F). P(E,F) seria a probabilidade de estar na faixa dos 30 anos e comprar algo, considerando a população total — não apenas a população de pessoas na faixa dos 30 anos.

```python
print("P(30's, Purchase)" + str(float(purchases[30]) / 100000.0))
```

Vamos também calcular o produto de P(E) e P(F), P(E)P(F):

```python
print("P(30's)P(Purchase)" + str(PE * PF))
```

Algo que você pode aprender em estatística é que P(E,F) = P(E)P(F), mas isso pressupõe que E e F sejam independentes. Descobrimos aqui que P(E,F) é aproximadamente 0,05, enquanto P(E)P(F) é aproximadamente 0,075. Portanto, quando E e F são dependentes — e temos uma probabilidade condicional envolvida — não podemos simplesmente dizer que P(E,F) = P(E)P(F).

Podemos também verificar que P(E|F) = P(E,F)/P(F), que é a relação que mostramos nos slides — e, de fato, é:

```python
print((purchases[30] / 100000.0) / PF)
```

Modifique o código acima de forma que a probabilidade de compra NÃO varie com a idade, tornando E e F de fato independentes.

Em seguida, confirme que P(E|F) é aproximadamente igual a P(E), mostrando que a probabilidade condicional de compra para uma determinada idade não difere da probabilidade a priori de compra, independentemente da idade.

```python
# Primeiro muda o código para ter uma probabilidade de compra fixa, independente da idade
random.seed(0)

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchases = 0
for _ in range(100000):
    ageDecade = random.choice([20, 30, 40, 50, 60, 70])
    purchaseProbability = 0.4 # Adicionei uns 40%
    totals[ageDecade] += 1
    if (random.random() < purchaseProbability):
        totalPurchases += 1
        purchases[ageDecade] += 1

# Em seguida, calcula P(E|F) para uma determinada faixa etária; escolhendo pessoas de 30 anos:
PEF = float(purchases[30]) / float(totals[30])
print("P(purchase | 30s): " + str(PEF))

# Agora cálcula o P(E)
PE = float(totalPurchases) / 100000.0
print("P(Purchase):" + str(PE))
```

```python
# Criando um dataframe com os dados do csv
df = pd.read_csv("vgsales_pratica.csv")
df.head()
```

```python
print(f"Mediana das vendas globais: {np.median(df['Global_Sales'])}")
print(f"Console com mais jogos lançados: {df['Platform'].mode()[0]}")
```

```python
# Mostra a média de vendas global e o desvio padrão das vendas
print("Média Global:", df['Global_Sales'].mean())
print("Desvio Padrão:", df['Global_Sales'].std())
```

```python
values = [np.sum(df['NA_Sales']), np.sum(df['EU_Sales']), np.sum(df['JP_Sales'])]
colors = ['r', 'g', 'b']
labels = ['América do Norte', 'Europa', 'Japão']
plt.pie(values, colors=colors, labels=labels)
plt.title('Vendas totais por região')
plt.show()
```

```python
# Momentos, a assimetria e a curtose.
print("Assimetria:", df['Global_Sales'].skew())
print("Curtose:", df['Global_Sales'].kurt())
# Assimetria maior que 0 indica que a cauda longa fica à direita, é o padrão de vendas de produtos
```

```python
# O kde=True desenha a linha da PDF
sns.histplot(df['Global_Sales'], kde=True, bins=100)
plt.title('Distribuição de Vendas Globais')
plt.xlim(0, 5) # Adicionei um limite para o gráfico não ficar achatado pelos outliers
plt.show()
```

```python
# Define o evento B
limite_hit = 10.0
df['Is_Hit'] = df['Global_Sales'] > limite_hit
# P(A): Qual a chance de um jogo aleatório ser Nintendo?
prob_nintendo = len(df[df['Publisher'] == 'Nintendo']) / len(df)
# P(B): Qual a chance de uma jogo ser um Hit?
prob_hit = len(df[df['Is_Hit'] == True]) / len(df)
# P(B|A): Dado que é Nintendo, qual a chance de ser Hit?
nintendo_games = df[df['Publisher'] == 'Nintendo']
prob_hit_dado_nintendo = len(nintendo_games[nintendo_games['Is_Hit'] == True]) / len(nintendo_games)
# Acha P(A|B): Dado que é Hit, qual a chance de ser Nintendo?
prob_nintendo_dado_hit = (prob_hit_dado_nintendo * prob_nintendo) / prob_hit

print(f"Probabilidade de ser Nintendo dado que é um Hit: {prob_nintendo_dado_hit}")

# Tirando a prova nos dados filtrados
hits = df[df['Is_Hit'] == True]
print(f"Prova reals dos dados filtrados: {len(hits[hits['Publisher'] == 'Nintendo']) / len(hits)}")
```
